# Sentiment Analysis

```python
import tensorflow_datasets as tfds

imdb, info = tfds.load('imdb_reviews', with_info=True, as_supervised=True)
```

```python
import numpy as np
train_data, test_data = imdb["train"], imdb["test"]

train_sentences = []
test_sentences = []

train_labels = []
test_labels = []

for sentence, label in train_data:
  train_sentences.append(str(sentence.numpy()))
  train_labels.append(label.numpy())

for sentence, label in test_data:
  test_sentences.append(str(sentence.numpy()))
  test_labels.append(label.numpy())
```

```python
train_sentences[0], train_labels[0]
```

```python
sample = train_sentences[0]
sample = sample.split()
print(len(sample))
sample[:10]
```

```python
train_labels = np.array(train_labels)
test_labels = np.array(test_labels)

train_labels.shape, test_labels.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = 10000
embedding_dim = 64
max_length = 140

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index
word_index
```

```python
train_sequences = tokenizer.texts_to_sequences(train_sentences)
padded_train_sequences = pad_sequences(train_sequences, maxlen=max_length, truncating="post", padding="post")
print(padded_train_sequences.shape)
padded_train_sequences
```

```python
test_sequences = tokenizer.texts_to_sequences(test_sentences)
padded_test_sequences = pad_sequences(test_sequences, maxlen=max_length, truncating="post", padding="post")
print(padded_test_sequences.shape)
padded_test_sequences
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dropout, Dense, Flatten

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length),
    Flatten(),
    Dense(10, activation="relu"),
    Dense(1, activation="sigmoid"),
])

model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(padded_train_sequences, train_labels, epochs=2, validation_data=(padded_test_sequences, test_labels))
```

```python
infer_text = ["I hate this movie very much, never watch it again"]

text_seq = tokenizer.texts_to_sequences(infer_text)
padded_text_seq = pad_sequences(text_seq, maxlen=max_length, truncating="post", padding="post")
padded_text_seq
```

```python
threshold = 0.5
labels = ['negative', 'positive']
prediction = model.predict(padded_text_seq)
prediction = 0 if prediction < threshold else 1
print(labels[prediction])
```

```python
embedding_layer = model.layers[0]
weights = embedding_layer.get_weights()[0]
weights
```

```python
import io
out_v = io.open('vecs.tsv', 'w', encoding='utf-8')
out_m = io.open('meta.tsv', 'w', encoding='utf-8')

for idx in range(1, vocab_size):
  word = tokenizer.index_word[idx]
  vector = weights[idx]
  out_m.write(word + "\n")
  out_v.write("\t".join([str(x) for x in vector]) + "\n")

out_m.close()
out_v.close()
```
